# Salary Prediction

```python
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
import matplotlib.pyplot as plt
sns.set()
```

## Loading the data

```python
data=pd.read_csv("Salary_dataset.csv")
data
```

```python
data.drop("Unnamed: 0",axis=1)
```

## Splitting the data

```python
y=data["Salary"]
x1=data["YearsExperience"]
```

## Plotting the data

```python
plt.scatter(x1,y)
plt.xlabel("Years")
plt.ylabel("Salary")
```

## Creating BestFit using OLS

```python
x=sm.add_constant(x1)
result=sm.OLS(y,x).fit()
result.summary()
```

## Creating regression line

```python
plt.scatter(x1,y)
yhat=24850+x1*9449.9623
plt.plot(x1,yhat,lw=4,color="red")
plt.xlabel("Years")
plt.ylabel("Salary")
```

## Prediction Function

```python
def predict(test):
    return 24850+test*9449.9623
test=int(input("Enter the years : "))
print(f"The salary is {predict(test)}")
```
