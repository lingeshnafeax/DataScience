# Dummy Variable

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
sns.set()
```

## Loading the data

```python
data=pd.read_csv("29588090-1.03.Dummies.csv")
data
```

## Creating Dummy Variable

```python
data["Attendance"]=data["Attendance"].map({"Yes":1,"No":0})
data
```

```python
data.describe()
```

## Splitting the data

```python
y=data["GPA"]
x1=data[["SAT","Attendance"]]
```

```python
x=sm.add_constant(x1)
result=sm.OLS(y,x).fit()
result.summary()
```

```python
sns.scatterplot(x=x1["SAT"],y=y,hue=data["Attendance"],style=data["Attendance"])
yhat=0.0017*x1["SAT"]+0.275                                                        #Original Regression line with only SAT
yhat_yes=0.6439+0.2226+0.0014*data["SAT"]                                          #Regression line with attendance = yes
yhat_no=0.6439+0.0014*data["SAT"]                                                  #Regression line with attendance = no
plt.plot(x1["SAT"],yhat,c="yellow")
plt.plot(x1["SAT"],yhat_yes,c="red")
plt.plot(x1["SAT"],yhat_no,c="blue")
```
