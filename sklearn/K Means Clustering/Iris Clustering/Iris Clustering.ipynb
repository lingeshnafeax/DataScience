# Iris Dataset

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
sns.set()
```

## load the dataset

```python
data=pd.read_csv("15453055-iris-dataset.csv")
data
```

```python
data.isna().sum()
```

## Plot the data

```python
plt.scatter(x=data["petal_length"],y=data["petal_width"])
plt.xlabel("Petal Lenght")
plt.ylabel("Petal Width")
```

## Standardising the x data

```python
from sklearn import preprocessing
data_scaled=preprocessing.scale(data)
data_scaled
```

```python
test=data_scaled[:,2:4]
test
```

## Use elbow method

```python
wcss=[]
for i in range(1,10):
    kmeans_temp=KMeans(i)
    kmeans_temp.fit(test)
    wcss_temp=kmeans_temp.fit_predict(test)
    wcss.append(kmeans_temp.inertia_)
```

```python
plt.plot(range(1,10),wcss)
plt.xlabel("Clusters")
plt.ylabel("WCSS")
```

Here 3 seems to be a better cluster value

```python
kmeans_new=KMeans(3)
kmeans_new.fit(test)
clusters=kmeans_new.fit_predict(test)
```

```python
df=pd.DataFrame(test)
df=df.rename(columns={0:"Pedal Length",1:"Pedal Width"})
df
```

```python
df["Clusters"]=clusters
df
```

## Plot the data

```python
plt.scatter(x=df["Pedal Length"],y=df["Pedal Width"],c=df["Clusters"],cmap="rainbow")
plt.xlabel("Pedal Length")
plt.ylabel("Pedal Width")
```

Thus we have found 3 clusters
