# Country Clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Loading the data

```python
data=pd.read_csv("29588950-Countries-exercise.csv")
data
```

## Plot the Data

```python
plt.scatter(data["Longitude"],data["Latitude"])
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```

## Select the features

```python
x=data.iloc[:,1:]
x
```

## Create cluster

```python
kmeans=KMeans(3)
```

```python
kmeans.fit(x)
```

```python
kmeans.get_params(deep=True)
```

```python
identified_clusters=kmeans.fit_predict(x)
identified_clusters
```

```python
data["Clusters"]=identified_clusters
data
```

```python
plt.scatter(data["Longitude"],data["Latitude"],c=data["Clusters"],cmap="rainbow")
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```
