```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from numpy.random import randint
%matplotlib inline
```

## LinePlot

```python
x=np.linspace(1,10,20)
x
```

```python
y=randint(1,50,20)
y
```

```python
plt.plot(x,y)
```

```python
y_cum=y.cumsum()
```

```python
plt.plot(x,y_cum,c="g",marker="o",linestyle="--")
```

```python
plt.plot(x,y_cum,c="g",marker="o",linestyle="--")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Relation")
plt.show()
```

## Scatter Plot

```python
x,y
```

```python
plt.scatter(x,y,linewidths=1)
plt.show()
```

##  Bar Chart

```python
a=[10,20,90,15,47]
b=["a","b","c","d","e"]
plt.bar(b,a,width=0.9)
plt.show()
```

```python
plt.hist(y,rwidth=0.8,bins=5)
plt.show()
```

## Box Plot

```python
plt.boxplot(y,vert=True,patch_artist=True)
plt.show()
```

## Sub Plot

```python
x,y_cum
```

```python
y2=x*y_cum
```

```python
plt.plot(x,y)
plt.plot(x,y2)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y_cum,c="r",marker="o")
plt.subplot(1,2,2)
plt.plot(x,y2,c="b",marker="*")
```

```python
fig,ax=plt.subplots(1,2,figsize=(12,4))
ax[0].plot(x,y_cum)
ax[0].set_xticks([1,5,10])
ax[0].set_xticklabels(["a","b","c"])
ax[1].plot(x,y2)
```

## Pie Chart

```python
x=[10,40,30,18,20]
labels=["Nexy","Nafeax","Loves","1975","So Much"]
plt.pie(x,labels=labels,autopct="%1.1f%%",startangle=360,explode=[0.1,0.1,0,0,0])
plt.show()
```
