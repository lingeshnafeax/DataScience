```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
%matplotlib inline
```

```python
data=sns.load_dataset("tips")
data.head()
```

```python
sns.relplot(data=data,x="total_bill",y="tip",col="smoker",hue="sex",kind="scatter",size="size")
plt.show()
```

```python
data=sns.load_dataset("diamonds")
data.head()
```

```python
sns.pairplot(data=data.filter(items=["carat","depth","table","price"]))
```
