```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
data=sns.load_dataset("tips")
data.head()
```

```python
sns.relplot(data=data,x="total_bill",y="tip",col="smoker",hue="sex",kind="scatter",size="size")
plt.show()
```

```python
data=sns.load_dataset("fmri")
data.head()
```

```python
sns.lineplot(data=data,x="timepoint",y="signal",hue="region",err_style="bars")
```
