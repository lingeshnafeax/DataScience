# EDA on Titanic Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set_style("whitegrid")
```

## Loading the data

```python
data=pd.read_csv("titanic_train.csv")
data.head()
```

```python
data.info(memory_usage="deep")
```

## Preprocessing

### Check for Missing Values

```python
data.isna().sum()
```

Cabin has 687 missing values
Age has 177 missing
Embaarked only has 2 missing values

We will drop 'Cabin'

```python
data=data.drop("Cabin",axis=1)
```

We will fill the missing 'Age' by median

```python
data["Age"]=data["Age"].fillna(data["Age"].median())
data
```

We will fill the missing 'Embarked' by mode

```python
data["Embarked"].fillna(data["Embarked"].mode()[0],inplace=True)
```

### 1.What is the total Count of Survivals An Victims?

```python
data["Survived"].value_counts()
```

```python
temp=data["Survived"].value_counts()
temp
```

```python
ax=sns.countplot(data=data,x="Survived")
ax.set_xticklabels(["Died","Survived"])
plt.title("Survival")
plt.xlabel("")
```

By this more people died than survived

### 2.Which Gender has more Survival Rate?

```python
data.groupby(["Sex","Survived"]).agg({"Survived":"count"})
```

```python
data["Survived"].describe()
```

```python
ax=sns.countplot(data=data,x="Survived",hue="Sex")
ax.set_xticklabels(["Died","Survived"])
```

We can say more women survied and more men died

```python
temp=data.groupby("Sex",as_index=False).agg({"Survived":"mean"})
temp
```

```python
sns.barplot(data=temp,x="Sex",y="Survived")
```

We conclude that survivors are more female than male

### 3.What is Survival rate based on Person type(Infant,Children,Adult)

```python
data.info()
```

```python
condition=[data["Age"] <= 1, (data["Age"]>1) & (data["Age"]<=12), data["Age"] >= 12]
choices=["Infant","Children","Adult"]
data["Age_band"]=np.select(condition,choices)     # Based of age we turned people into 3 category
```

```python
data["Age_band"].value_counts()
```

```python
data.groupby(["Survived","Age_band"]).agg({"Survived":"count"})
```

1. Adult Survival Rate

```python
ax=sns.countplot(data=data[data["Age_band"]=="Adult"],x="Survived")
ax.set_xticklabels(["Died","Survived"])
```

```python
plt.pie(data[data["Age_band"]=="Adult"].loc[:,"Survived"].value_counts(), labels = ["Survived","Dead"], autopct='%.1f%%')
plt.title("Adult Survival Rate")
plt.show()
```

More adults died than surviving

2. Children Survival Rate

```python
plt.pie(data[data["Age_band"]=="Children"].loc[:,"Survived"].value_counts(), labels = ["Survived","Dead"], autopct='%.1f%%')
plt.title("Children Survival Rate")
plt.show()
```

Half children survived and half dead

3. Infant Survival Rate

```python
plt.pie(data[data["Age_band"]=="Infant"].loc[:,"Survived"].value_counts(), labels = ["Survived","Dead"], autopct='%.1f%%')
plt.title("Infant Survival Rate")
plt.show()
```

Almost 85% of the infants died
